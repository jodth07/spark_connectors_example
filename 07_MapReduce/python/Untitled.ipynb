```python
base_dir = f"{os.getcwd()}/00_input/data"
```

```python
bible_path = f"{base_dir}/bible-kjv.txt"
shakes_path = f"{base_dir}/shakespeare.txt"
```

```python
bible_lines = open(bible_path, "r", encoding="utf-8").read().lower()
shakes_lines = open(shakes_path, "r", encoding="utf-8").read().lower()
```

```python
punc = string.punctuation + "\n"
stop_words = stopwords.words("english")
```

```python
def get_word_count(text: str):
    clean_bible = bible_lines.translate(str.maketrans(punc, " "*len(punc)))
    wc_dict = {}
    for word in text.split():
        if word not in stop_words and not word.isnumeric():
            if word not in wc_dict:
                wc_dict[word] = 1
            else :
                wc_dict[word] += 1
    order_dict = {k: v for k, v in sorted(wc_dict.items(), key=lambda item:item[1], reverse=True)}
    return order_dict
```

```python
bible_wc = get_word_count(bible_lines)
```

```python
shake_wc = get_word_count(shakes_lines)
```

```python
# words, bible, shakes
```

```python
bible_100 = dict(list(bible_wc.items())[: 100]) 
```

```python
shake_100 = dict(list(shake_wc.items())[: 100]) 
```

```python
words_freq_df = pd.DataFrame([bible_100, shake_100]).transpose().reset_index()
```
